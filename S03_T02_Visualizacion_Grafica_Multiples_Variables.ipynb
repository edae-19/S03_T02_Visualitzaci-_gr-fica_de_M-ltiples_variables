```python
# Data to be used

x = tips_df['total_bill']
y = tips_df['tip']

# Plot - Scatter plot - Tips per Total bill - Customer sex

plt.figure (figsize = (10,5))

sns.scatterplot (data = tips_df, x = x, y = y, hue = 'sex')

fuente_titulo = {'family' : 'DejaVu Sans', 'color' : 'steelblue', 'size' : 20}
fuente_ejes = {'family' : 'DejaVu Sans', 'color' : 'black', 'size' : 12}

plt.title("Tips vs. Total Bill - Sex distribution", fontdict = fuente_titulo)
plt.xlabel("Total bill (€)", fontdict = fuente_ejes)
plt.ylabel("Tips (€)", fontdict = fuente_ejes)

plt.show()

print ('CONCLUSION\nThere is an apparent linear relationship between the total bill and the tip. The higher one the higher the other\nMen tend to spend the highest amount ')
```

```python
#Data to be used
x = tips_df['tip']

# Plot - Histogram - Tips

plt.figure (figsize = (10,5))

sns.histplot(data = x, binwidth = 0.5, kde = True, color ="olivedrab")

fuente_titulo = {'family' : 'DejaVu Sans', 'color' : 'steelblue', 'size' : 20}
fuente_ejes = {'family' : 'DejaVu Sans', 'color' : 'black', 'size' : 12}

plt.title("Tips Distribution", fontdict = fuente_titulo)
plt.xlabel("Tip amount (€)", fontdict = fuente_ejes)
plt.ylabel("Count of tips (unt)", fontdict = fuente_ejes)

plt.show()
```

```python
# Average - STD:
mean = round(tips_df.tip.mean(),3)
std_tip = round(tips_df.tip.std(),3)

percentile_25 = round(tips_df['tip'].quantile(0.25),3)
percentile_75 = round (tips_df['tip'].quantile(0.75),3)
median = tips_df['tip'].median()
```

### EJERCICIO 2 

```python
movies_df = pd.read_csv('movies.txt',sep ='::', engine = 'python', header = None)
```

```python
movies_df.head()
```

```python
movies_df.rename(columns = {0:'Index', 1:'Movie',2:'Gender'}, inplace = True)
```

```python
movies_df.info()
```

```python
# PROCESING 2nd COLUMN

gender_splitted_df = movies_df['Gender'].str.split('|', expand = True)
gender_splitted_df.info()
```

```python
gender_splitted_df.head()
```

```python
movies_df['gender1'] = gender_splitted_df[0]
movies_df['gender2'] = gender_splitted_df[1]
movies_df['gender3'] = gender_splitted_df[2]
movies_df['gender4'] = gender_splitted_df[3]
movies_df['gender5'] = gender_splitted_df[4]
movies_df['gender6'] = gender_splitted_df[5]
```

```python
gender_splitted_df.shape[1]
```

```python
movies_df.head()
```

```python
# GENERATE A LIST WITH ALL VALUES IN THE GENDER COLUMN 
gender_list = []

for i in range (gender_splitted_df.shape[1]):
    list_1 = gender_splitted_df[i].tolist() 
    gender_list = gender_list + list_1
      
```

```python
gender_type = []

for gender in gender_list:
    if gender not in gender_type:
        gender_type.append(gender)
        
gender_type.remove(None)
```

```python
# PROCESING 1st COLUMN
years = []

for movie in movies_df['Movie']:
    year = movie[-5:-1]
    years.append(year)

movies_df ['Movie_Year'] = years
```

```python
movies_df.head()
```

```python
#PLOT 1 - EVOLUTION OF MOVIES PER YEAR

movie_year_df = movies_df.groupby (['Movie_Year'], as_index = False).count()
```

```python
plt.figure (figsize = (16,7))

sns.lineplot(data = movie_year_df, x = movie_year_df['Movie_Year'], y = movie_year_df['Movie'])

fuente_titulo = {'family' : 'DejaVu Sans', 'color' : 'steelblue', 'size' : 20}
fuente_ejes = {'family' : 'DejaVu Sans', 'color' : 'black', 'size' : 12}

plt.title("MOVIES PRODUCTION PER YEAR", fontdict = fuente_titulo)
plt.xlabel("Year", fontdict = fuente_ejes)
plt.xticks(rotation=90)
plt.ylabel("Number of movies (Unt)", fontdict = fuente_ejes)

plt.show()

print ('CONCLUION\nThere was a sharp increase in movies production between 1991 and 1995 by almost tripling the average movies per year of the previous years.')
```

```python
# PLOT 2 -  MOVIES PER GENDER

count_movies_gender = []

for gender in gender_type:
    counter = 0
    for i in range (movies_df.shape[0]):
        if gender in movies_df.iloc[i]['Gender']:
            counter +=1
    count_movies_gender.append(counter)
    
```

```python
gender_df = pd.DataFrame (list(zip(gender_type,count_movies_gender)), columns = ['Gender', 'Count'])
gender_df = gender_df.sort_values ('Count', ascending = False )
```

```python
plt.figure (figsize = (16,7))

sns.barplot(data = gender_df, x = 'Gender', y = 'Count', palette="viridis")

fuente_titulo = {'family' : 'DejaVu Sans', 'color' : 'steelblue', 'size' : 20}
fuente_ejes = {'family' : 'DejaVu Sans', 'color' : 'black', 'size' : 12}

plt.title("MOST POPULAR GENDER", fontdict = fuente_titulo)
plt.xlabel("Gender", fontdict = fuente_ejes)
plt.xticks(rotation=45)
plt.ylabel("Movies", fontdict = fuente_ejes)

plt.show()

print ('CONCLUION\nAs we can see in the graphic, \'Drama\' and \'Comedy\' are, by far, the most popular genders within the studied movies')
```

```python
# PLOT 3 - EVOLUTION OF MOVIES PER GENDER

gender1_df = movies_df.loc[:,('Movie_Year','Movie','gender1')]
gender1_df.rename (columns = {'gender1' : 'Gender'}, inplace = True)

gender2_df = movies_df.loc[:,('Movie_Year','Movie','gender2')]
gender2_df.rename (columns = {'gender2' : 'Gender'}, inplace = True)

gender3_df = movies_df.loc[:,('Movie_Year','Movie','gender3')]
gender3_df.rename (columns = {'gender3' : 'Gender'}, inplace = True)

gender4_df = movies_df.loc[:,('Movie_Year','Movie','gender4')]
gender4_df.rename (columns = {'gender4' : 'Gender'}, inplace = True)

gender5_df = movies_df.loc[:,('Movie_Year','Movie','gender5')]
gender5_df.rename (columns = {'gender5' : 'Gender'}, inplace = True)

gender6_df = movies_df.loc[:,('Movie_Year','Movie','gender6')]
gender6_df.rename (columns = {'gender6' : 'Gender'}, inplace = True)
```

```python
year_gender_df = gender1_df.append (gender2_df)

year_gender_df = year_gender_df.append(gender3_df)

year_gender_df = year_gender_df.append(gender4_df)

year_gender_df = year_gender_df.append(gender5_df)

year_gender_df = year_gender_df.append(gender6_df)
```

```python
year_gender_df.dropna(inplace = True)
```

```python
df =year_gender_df.groupby (['Movie_Year','Gender'], as_index =False).count()
```

```python
plt.figure (figsize = (16,7))

sns.lineplot(data = df, x = df['Movie_Year'], y = df['Movie'], hue = 'Gender' )

fuente_titulo = {'family' : 'DejaVu Sans', 'color' : 'steelblue', 'size' : 20}
fuente_ejes = {'family' : 'DejaVu Sans', 'color' : 'black', 'size' : 12}

plt.title("Evolution of Movies Gender per Year", fontdict = fuente_titulo)
plt.xlabel("Year", fontdict = fuente_ejes)
plt.xticks(rotation=90)
    #plt.xlim (1919,1930)
plt.ylabel("Number of movies (Unt)", fontdict = fuente_ejes)

plt.show()

print ('CONCLUSION\nEven though we could clearly see a dominance in the Drama an Comedy movies, we can not easily view the evolution of the rest of genders')
```
